```python
def f(a):
    if (a>0 and a<=10):
        return '0 to 10'
    elif (a>10 and a<=20):
        return '10 to 20'
    elif (a > 20 and a <= 30):
        return '21 to 30'
    elif (a > 30 and a <= 40):
        return '31 to 40'
    elif (a > 40 and a <= 50):
        return '41 to 50'
    elif (a > 50 and a <= 60):
        return '51 to 60'
    elif (a > 60 and a <= 70):
        return '61 to 70'
    elif (a > 70 and a <= 80):
        return '71 to 80'
    elif (a > 80 and a <= 90):
        return '81 to 90'
    elif (a > 90 and a <= 100):
        return '91 to 100'
    else:
        return 'Out of Range'



    
f(50)
```
